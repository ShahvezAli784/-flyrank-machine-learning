## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

**Task Type:** Scoring (Regression)

My lane is a scoring problem. The goal is to estimate how much a piece of content needs to be refreshed based on its search performance, engagement, age, and freshness. Instead of assigning categories, the model predicts a numeric score that helps prioritize which content should be updated first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*


**Target (Proxy):** trend_pct

The dataset does not contain a direct label indicating whether content should be refreshed. Instead, I use **trend_pct** as a proxy target because it measures how content performance is changing over time. A negative trend suggests declining performance and may indicate that the content should be refreshed.

## 3. Success metric

*One metric you can defend. What number means 'good'?*


**Metric:** Mean Absolute Error (MAE)

I will evaluate the model using Mean Absolute Error (MAE). It measures the average difference between the predicted trend percentage and the actual trend percentage. Lower MAE indicates better predictions. A lower prediction error means the model can more accurately prioritize which content needs attention.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*


Content performance depends on many interacting factors such as search volume, competition, click-through rate, average position, engagement rate, content age, and user behavior. These relationships are complex and difficult to capture using simple if-else rules. A machine learning model can learn patterns from historical data and provide more accurate refresh recommendations than a fixed rule-based approach.
