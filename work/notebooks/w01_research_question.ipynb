## 3. Quick look at the data (2-3 real numbers)

```python
import pandas as pd

url = "https://raw.githubusercontent.com/ShahvezAli784/-flyrank-machine-learning/main/data/raw/content_refresh_anonymized.csv"

df = pd.read_csv(url)

df.head()
```

```python
# Number of rows and columns
print("Dataset shape:", df.shape)

# Declining pages
print("\nTrend direction:")
print(df["trend_direction"].value_counts())

# Percentage of declining pages
declining_pct = (df["trend_direction"] == "down").mean() * 100
print(f"\nDeclining pages: {declining_pct:.2f}%")

# Median impressions
print("\nMedian impressions (90d):", df["impressions_90d"].median())

# Average CTR
print("Average CTR:", df["ctr"].mean())

# Average content age
print("Average content age (days):", df["content_age_days"].mean())
```
